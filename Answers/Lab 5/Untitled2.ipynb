```python
latitude = list(todosd['latitude'])
longitude = list(todosd['longitude'])
last_seen = list(todosd['last_seen'])
brand = list(todosd['brand'])
for latitude,longitude,last_seen,brand in zip(latitude,longitude,last_seen,brand):
    test = folium.Html('<b>{{last_seen}}<br>{brand}</b>', script=True) # i'm assuming this bit runs fine
    iframe = branca.element.IFrame(html=test, width=350, height=150)
    popup = folium.Popup(iframe, parse_html=True)
    folium.Marker(location=[latitude,longitude],radius=6,color='grey',fill_color='yellow', popup=popup, icon= icon).add_to(map)
```
